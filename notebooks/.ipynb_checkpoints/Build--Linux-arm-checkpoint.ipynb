```bash
BUILD_DIR=/workdir/build
SOURCE_DIR=$BUILD_DIR/icu4c-source
TARGET_DIR=$BUILD_DIR/icu4c-target
LIB_DIR=$(zig env | jq -r '.lib_dir')
```

```bash
rm -Rf "$TARGET_DIR"
```

```bash
rm -Rf "$TARGET_DIR"/linux-arm-64
rm -Rf "$TARGET_DIR"/linux-x86-64
make distclean || make clean
```

```bash
OS=linux
ARC=x86_64

# paths
HOST_BUILD=$BUILD_DIR/icu4c-build-host        # <— the host *build* dir (NEW)
HOST_INSTALL=$TARGET_DIR/${OS}-${ARC}         # <— the host *install* dir (your prefix)

LIB_DIR=$(zig env | jq -r '.lib_dir')
SYSROOT="$LIB_DIR/libc"

rm -rf "$HOST_BUILD" && mkdir -p "$HOST_BUILD"
mkdir -p "$HOST_INSTALL"
cd "$HOST_BUILD"

export CC="zig cc -target x86_64-linux-musl --sysroot $SYSROOT"
export CXX="zig c++ -target x86_64-linux-musl --sysroot $SYSROOT"
export CXXFLAGS="-std=c++20"

"$SOURCE_DIR"/source/configure        \
  --prefix="$HOST_INSTALL"     \
  --enable-static              \
  --disable-shared             \
  --with-data-packaging=static \
  --disable-samples            \
  --disable-tests

make -j"$(nproc)"
make install
```

```bash
OS=linux
ARC=arm_64

ARM_BUILD=$BUILD_DIR/icu4c-build-${OS}-${ARC}
ARM_INSTALL=$TARGET_DIR/${OS}-${ARC}
ZIG_TARGET="aarch64-linux-musl"

rm -rf "$ARM_BUILD" && mkdir -p "$ARM_BUILD"
mkdir -p "$ARM_INSTALL"
cd "$ARM_BUILD"

export CC="zig cc -target $ZIG_TARGET --sysroot $SYSROOT"
export CXX="zig c++ -target $ZIG_TARGET --sysroot $SYSROOT"
export CXXFLAGS="-std=c++20"

"$SOURCE_DIR"/source/configure     \
  --host=$ZIG_TARGET               \
  --with-cross-build="$HOST_BUILD" \
  --prefix="$ARM_INSTALL"          \
  --enable-static                  \
  --disable-shared                 \
  --with-data-packaging=static     \
  --disable-samples                \
  --disable-tests

make -j"$(nproc)"
make install
```

```bash
for p in linux-x86_64 linux-arm_64 ; do
    p_PATH=/workdir/build/icu4c-target/$p
    echo $p_PATH     ; ls -la $p_PATH     ; echo 
    echo $p_PATH/lib ; ls -la $p_PATH/lib ; echo 
    for f in $(ls /workdir/build/icu4c-target/$p/lib/*.a) ; do
        echo "$p"
        file $f
        
        echo -n "$f: "
        first_obj=$(ar t "$f" | head -n 1)
        ar p "$f" "$first_obj" | file - | sed 's/^ *//'
        
        echo 
        
        echo -n "$p"
        nm $f | head -n 10
        if [ $(nm "$f" | wc -l) -gt 10 ]; then
          echo "..."
        fi
        echo 
    done
    echo
done
echo 
```
